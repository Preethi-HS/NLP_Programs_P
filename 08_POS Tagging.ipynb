Universal Tagset, a standardized set of labels used to classify parts of speech in text.
Downloading the Universal Tagset is necessary if you want to use NLTK's Part-of-Speech (POS) tagging functionality with the Universal Tagset. POS tagging involves assigning a label (e.g., noun, verb, adjective) to each word in a sentence to understand its grammatical function.


```python
import nltk
nltk.download("universal_tagset")
```

The downloaded model helps NLTK assign grammatical labels (e.g., noun, verb, adjective) to words in a sentence, a process called POS tagging.
Benefits: Using a pre-trained model like the Averaged Perceptron tagger can offer higher accuracy and efficiency compared to simpler rule-based taggers.


```python
nltk.download("averaged_perceptron_tagger")
```

```python
from nltk.tokenize import TreebankWordTokenizer
from nltk import pos_tag


twt = TreebankWordTokenizer()
tags=pos_tag(twt.tokenize(text),tagset='universal')
tags

```

```python
spans=[i for i in twt.span_tokenize(text)]
spans
#strating and ending index
```

a list of tuples, where each tuple represents a word and its part-of-speech tag. For example, `[("This", "DT"), ("is", "VBZ"), ("a", "DT"), ("sample", "NN"), ("sentence", "NN")]`





```python
import random
pos=list(set(i[1] for i in tags))
colors=dict.fromkeys(pos)
#a method that creates a dictionary with all the keys from the pos list
print(colors)
for i in colors.keys():
    colors[i] =" ".join(random.sample(['salmon','khaki','blue','orange','red','green'],k=1))
print(colors)
ents=[] #list of dict
#where each dictionary represents an entity (word) along with its starting index, ending index, and part-of-speech label.
for tag, span in zip(tags,spans):
    ents.append({"start":span[0],'end':span[1],'label':tag[1]})

ents
```

```python
doc={"text":text,"ents":ents}
options={"ents":pos,"colors":colors}
displacy.render(doc,style="ent",options=options,manual=True)
```
