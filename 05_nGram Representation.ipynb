## 4. Generate n-grams(bi-grams)/trigrams and freq.

```python
def ngram(k):
    l = []
    for i in range(len(token)):
        l.append(token[i:i+k])
    l=l[:-1]

    # Get Freq.
    dct = {}
    l = [" ".join(i) for i in l]
    # for i in l:
    #     dct[i] = 0
    # for i in l:
    #     dct[i] += 1
    for i in l:
      dct[i]=dct.get(i,0)+1
    return dct
```
