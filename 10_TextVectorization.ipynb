```python
msg = ["Shaw likes to play cricket", "Mary likes to play tennis", "John likes to play volleyball or cricket", "Heena likes to play tennis or throwball"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd


vec = CountVectorizer()
bag = vec.fit_transform(msg).toarray()
print(vec.get_feature_names_out())
df = pd.DataFrame(bag, columns=[vec.get_feature_names_out()])
df

```

## Tf-IDF

```python
vec = TfidfVectorizer()
bag = vec.fit_transform(msg)
print(vec.get_feature_names_out())
df = pd.DataFrame(bag.toarray(), columns=vec.get_feature_names_out())
df
```

## Word2Vec
 Processing text data from a file called 'word2vec.txt', tokenizing it, and then training a Word2Vec model using the Gensim library.

Import necessary libraries:
- `nltk.tokenize` for tokenization.
- `re` for regular expressions.
- `nltk` for natural language processing tasks.
- `gensim.models.Word2Vec` for training Word2Vec models.

```python
from nltk.tokenize import word_tokenize
with open('word2vec.txt') as f:
  text = f.readlines()
text[0]
```

Use regular expressions to remove non-alphabetical characters from each line of text in the text list.

```python
import re
for i in range(len(text)):
  text[i] = re.sub("[^a-zA-Z]", ' ', text[i]) #substitute non-alphabetical characters in each line of text
print("Number of sentences: ", len(text))
print("First Sentence from the text file: \n", text[0])
```

Tokenize each sentence in the text data using `nltk.word_tokenize()` and store the tokens in a list called `token`.

```python
import nltk
nltk.download('punkt')
token = []
for i in text:
  token.append(word_tokenize(i))
token[0]
```

- Import `Word2Vec` from `gensim.models` and train a Continuous Bag of Words (CBOW) Word2Vec model (`cbow`) using the tokenized sentences. Parameters such as `min_count` and `vector_size` are set to 2 each.
- Retrieve the vocabulary (words) learned by the model using `cbow.wv.index_to_key`.
- Print the total number of words in the vocabulary.


```python
from gensim.models import Word2Vec
cbow = Word2Vec(sentences = token, min_count=2, vector_size = 2)
words = cbow.wv.index_to_key
words
```

Retrieve the vector representation of the word 'and' using `cbow.wv.__getitem__('and')`.
This method is used to retrieve the word vector for the specified word, which in this case is "and".



```python
print('Total Words: ', len(words))
cbow.wv.__getitem__('and')
```
