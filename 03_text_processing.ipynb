## Expand Contractions
 Contractions refer to shortened forms of words or phrases that are formed by combining two words into one. Contractions often involve replacing one or more letters with an apostrophe (') to indicate that letters have been omitted.

 Example:
- "can't" (short for "cannot")
- "won't" (short for "will not")
- "I'm" (short for "I am")

```python
import contractions
text=data['Email Content'][0] #Accesses the first element of the sequence obtained from data['Email Content']
print('original text:\n',text)
print('expanded text:\n')
for i in text.split(): #This loop iterates over each word in the original text (text) by splitting the text into a list of words using text.split()
    print(contractions.fix(i), end=' ')# =to expand
```

## Remove Punctuation

```python
import re
data['text']=data['Email Content'].apply(lambda x: re.sub(r'[^\w\s]','',x)) #match any which is not a word or white space
data['text'][0]
```

## Removing Digits

```python
data['text']=data['text'].apply(lambda x: re.sub('\w*\d\w*','',x)) #any freq of number
data['text'][3]
```

## Removing stop words and specified words

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words('english'))
stop_words.add('subject')
def remove_stopwords(text):
    return" ".join([word for word in str(text).split() if word not in stop_words])
data['text']=data['text'].apply(lambda x: remove_stopwords(x))
data['text'][3]
```
